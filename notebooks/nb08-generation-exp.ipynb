```python
def live_generate(mod, tok, text, write_fn, max_length, *skip_tokens,
                  min_length=0, chunk_size=20, pause=.05, **kwargs):
    for char in text:
        write_fn(char)
        time.sleep(pause)
    x = tok.encode(text, return_tensors='pt').squeeze()
    total_len = x.shape[-1] + max_length
    while len(x) < total_len:
        text, new, x = generate(mod, tok, text, None, *skip_tokens, 
                                max_length=min(chunk_size, total_len-len(x)), 
                                min_length=min_length, verbose=False, 
                                return_idx=True, **kwargs)
        for char in new:
            write_fn(char)
            time.sleep(pause)
        text += new
    return text
```

## Embedding Experiments

```python
from itertools import combinations
from scipy.spatial.distance import cosine, euclidean
from sklearn.decomposition import PCA

from incendio.nlp import Embeddings, Vocabulary
```

```python
@valuecheck
def distance(s1, s2, agg:('auto', 'cat', 'mean', 'sum')='sum', 
             mode:('cosine', 'euclidean')='cosine'):
    i1, i2 = tok.encode([s1]), tok.encode([s2])
    t1, t2 = emb[i1], emb[i2]
    fn = cosine if mode == 'cosine' else euclidean
    if agg in ('auto', 'cat'):
        if len(i1) == len(i2):
            t1, t2 = torch.cat([*t1], dim=-1), torch.cat([*t2], dim=-1)
        elif agg == 'auto':
            print('Tenshor shape mismatch. Falling back to agg=sum.')
            return distance(s1, s2, agg='sum', mode=mode)
        else:
            raise ValueError('Tensor shape mismatch.')
    elif agg in ('mean', 'sum'):
        agg_fn = partial(getattr(torch, agg), dim=0)
        t1, t2 = map(agg_fn, [t1, t2])
    return fn(t1, t2)
```

```python
def pairwise_distances(*terms, agg='sum'):
    scores = [{'t1': t1, 't2': t2, 
               'cosine': distance(t1, t2, agg, 'cosine'), 
               'euclidean': distance(t1, t2, agg, 'euclidean')}
              for t1, t2 in combinations(terms, 2)]
    return pd.DataFrame(scores)
```

```python
def w2vec(word, agg='sum'):
    idx = tok.encode([word])
    return getattr(torch, agg)(emb[idx], dim=0)
```

```python
def get_vocab(*terms, agg='sum'):
    w2v = {t: w2vec(t, agg) for t in terms}
    w2i = {t: i for i, t in enumerate(terms, 2)}
    vocab = Vocabulary(w2i, w2v, all_lower=False)
    return vocab
```

```python
def get_embeddings_analyzer(*terms, agg='sum'):
    mat = torch.stack([w2vec(t, agg) for t in terms], dim=0).numpy()
    w2i = {t: i for i, t in enumerate(terms)}
    return Embeddings(mat, w2i)
```

```python
def plot_3d(terms):
    x, y, z = map(list, zip(*[mat3d[e.w2i[t.lower()]] for t in terms]))
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.scatter(x, y, z)
    for i, t in enumerate(terms):
        ax.text(x[i], y[i], z[i], t)
    plt.show()
```

```python
def plot_2d(terms):
    x, y, z = map(list, zip(*[mat3d[e.w2i[t.lower()]] for t in terms]))
    for d1, d2 in combinations((x, y, z), 2):
        fig, ax = plt.subplots()
        ax.scatter(d1, d2, s=1)
        for i, t in enumerate(terms):
            ax.text(d1[i], d2[i], t, fontsize=8)
        plt.show()
```
